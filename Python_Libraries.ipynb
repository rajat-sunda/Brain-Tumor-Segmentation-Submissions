# Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Matplotlib

```python
# Generate data
np.random.seed(42)

normal_data = np.random.normal(loc=0, scale=1, size=1000)
beta_data = np.random.beta(a=2, b=5, size=1000)

# Create subplots
plt.figure(figsize=(10, 4))

# Beta distribution histogram
plt.subplot(1, 2, 1)
plt.hist(beta_data, bins=20, density=True, color = 'blue', edgecolor = "black")
plt.title("Beta Distribution (a=2, b=5)")
plt.xlabel("Value")
plt.ylabel("Density")

# Normal distribution histogram
plt.subplot(1, 2, 2)
plt.hist(normal_data, bins=20, density=True, color = 'green', edgecolor = 'black')
plt.title("Normal Distribution (mean = 0, var = 1)")
plt.xlabel("Value")
plt.ylabel("Density")

plt.tight_layout()
plt.show()
```

```python
np.random.seed(42)

# Generate normal distribution data
data = np.random.normal(loc=0, scale=1, size=1000)

# Compute statistics
mean = np.mean(data)
p25 = np.percentile(data, 25)
p75 = np.percentile(data, 75)

# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=20, density=True, color="blue", edgecolor="black", alpha=0.6)

# Vertical lines
plt.axvline(mean, color="black", linestyle="--", linewidth=2, label=f"Mean = {mean:.2f}")
plt.axvline(p25, color="red", linestyle=":", linewidth=2, label=f"25th Percentile = {p25:.2f}")
plt.axvline(p75, color="green", linestyle=":", linewidth=2, label=f"75th Percentile = {p75:.2f}")

# Annotations
plt.text(mean, 0.35, "Mean", color="black", ha="left")
plt.text(p25, 0.30, "25%", color="red", ha="right")
plt.text(p75, 0.30, "75%", color="green", ha="right")

# Titles and labels
plt.title("Histogram of Normal Distribution with mean zero and unit variance", fontsize=14)
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## Assignment 2

```python
poison_data = np.random.poisson(lam = 5, size = 20) # lambda = Decide the average number of events for your interval (e.g., 3 calls per minute).
std_dev = np.std(poison_data)

Normalised_poisson = (poison_data- np.mean(poison_data))/std_dev
print(f"Normalized Poisson data: {Normalised_poisson}")
```

## Assignment 1

```python
df = pd.read_csv("/content/videogamesales.csv")
df['global_sales'] = df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales'] + df['Other_Sales']
display(df.head())
```

```python
df_sorted = df.sort_values(by = 'global_sales', ascending = False)
display(df_sorted.head())
```

```python
copies_sold_per_genre = df.groupby('Genre')['global_sales'].sum().sort_values(ascending = False)
plt.bar(copies_sold_per_genre.index, copies_sold_per_genre.values, color = 'green', edgecolor = 'black')
plt.xlabel('Genre')
plt.ylabel('Total Sales')
plt.title('Total Sales per Genre')
plt.xticks(rotation=45)
plt.show()
```

```python
df3 = df[df['Name'] == 'Grand Theft Auto']
df3.head()
```

```python
lst = ["North America", "Europe", "Japan", "Other Sales"]
count = [df3.sum().NA_Sales, df3.sum().EU_Sales, df3.sum().JP_Sales, df3.sum().Other_Sales]
colors = ["red", "yellow", "lightgreen", "pink"]
plt.pie(count, labels = lst, colors = colors, autopct = '%1.1f%%', shadow = True)
plt.title("Total sales of all Grand Theft Auto game in different regions")
plt.show()
```
